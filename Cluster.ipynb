```python
iris = datasets.load_iris()
clust1 = scale(iris.data)
clust2 = scale(np.random.rand(150,4))
```

```python
inertia = []
for i in range(1,8):
  kmeans = KMeans(n_clusters=i, n_init='auto')
  kmeans.fit(clust1)
  inertia.append(kmeans.inertia_)
plt.plot(range(1,8), inertia)
plt.title('Elbow')
plt.show()
```

```python
inertia = []
for i in range(1,8):
  kmeans = KMeans(n_clusters=i, n_init='auto')
  kmeans.fit(clust2)
  inertia.append(kmeans.inertia_)
plt.plot(range(1,8), inertia)
plt.title('Elbow')
plt.show()
```
